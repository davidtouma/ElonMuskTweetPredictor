```python
elon_musk_tweet_df = pd.read_csv('cleanedTweetData.csv')
```

```python
elon_musk_tweet_df.head()
```

We can just drop the unnamed column

```python
elon_musk_tweet_df = elon_musk_tweet_df.drop(columns=['Unnamed: 0'])
```

```python
elon_musk_tweet_df.head()
```

Now we can upload the data to hopsworks

```python
#Training Feature pipeline
#Have your hopsworks API key ready

LOCAL = True

def g():
  feature_list = []
  
  project = hopsworks.login()
  fs = project.get_feature_store()
  
  for features in elon_musk_tweet_df.columns:
    feature_list.append(features)

  em_tweet_fg = fs.get_or_create_feature_group(
      name="elon_musk_tweets_modal",
      version=1,
      primary_key= feature_list,
      description = "Elon Musks Tweets from 2010-2022")
  em_tweet_fg.insert(elon_musk_tweet_df, write_options={"wait_for_job" : False})
  
  
if __name__ == "__main__":
  if LOCAL == True :
      g()
```

Now the data should be in Hopsworks :)
